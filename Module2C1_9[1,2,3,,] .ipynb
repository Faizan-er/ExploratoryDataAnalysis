# [C7] CDF: Cumulative distribution function


```python
#Some Prior Defintions   AAGE ESTEMAL HOGA ESKA CODE ME

iris_setosa = iris.loc[iris["Species"] == "setosa"];  
iris_virginica = iris.loc[iris["Species"] == "virginica"];
iris_versicolor = iris.loc[iris["Species"] == "versicolor"]; 
```

```python
#Need for CDF: We can visually see the %ge of Versiclor with PL less than 1.6
#Constructing CDF  #Reading CDF

#Computing PDF
counts,bin_edges = np.histogram(iris_setosa['PetalLengthCm'], bins=10,density= True)
                       #change thi bins = 9 , 11, 13 and See: study why ? How ?

pdf= counts/(sum(counts))
print(pdf);
print(bin_edges)

#comuting CDF

cdf= np.cumsum(pdf)                  #Cumulative sum of pdf till that point
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)

plt.show();
```

- Value of CDF always Lies in between [0,1] 
- see at 1.6: 82 % of Setosa flower lies before PL =1.6 Cm 
- Intigration of CDF till any point give CDF.
- Difrentiation of CDF--> PDF ------ at any point
- By looking at CDF of corrospondig pdf you can estimate the percentage the flower till the point of observation on x axis.

# [C8] Mean, Variance, Standard deviation  

```python
print(np.mean(iris_setosa["PetalLengthCm"])) 
print(np.mean(iris_virginica["PetalLengthCm"]))
print(np.mean(iris_versicolor["PetalLengthCm"]))

#mean with outlier: One Outlier datapoint can make Huge diffrence in the mean (Outlier may be data-fault, currupt)
print(np.mean(np.append(iris_setosa["PetalLengthCm"],50)))


 
print("\n Std-Dev")
print(np.std(iris_setosa["PetalLengthCm"]))
print(np.std(iris_virginica['PetalLengthCm']))
print(np.std(iris_virginica['PetalLengthCm']))
print(np.std(iris_versicolor["PetalLengthCm"]))
```
